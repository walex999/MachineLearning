```python
import scipy.io
import matplotlib.pyplot as plt

data = scipy.io.loadmat('data.mat')
train_X = data['X'].T
train_y = data['y'].T
test_X = data['Xval'].T
test_y = data['yval'].T

plt.scatter(train_X[0, :], train_X[1, :], c=train_y[0, :], s=40, cmap=plt.cm.Spectral);
```

# Design a neural network w/o regularization

```python
from keras import models
from keras import layers

model_noreg = models.Sequential()

model_noreg.add(layers.Dense(128, input_dim = 2, activation='relu'))
model_noreg.add(layers.Dense(128, activation='relu'))
model_noreg.add(layers.Dense(128, activation='relu'))
model_noreg.add(layers.Dense(128, activation='relu'))
model_noreg.add(layers.Dense(1, activation='sigmoid'))

model_noreg.compile(optimizer='rmsprop', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

history = model_noreg.fit(train_X.T, train_y.T, epochs=1000, batch_size=32,validation_data=(test_X.T,test_y.T))  # Adjust the number of epochs and batch size as needed
```

# Draw decision boundaries w/o regularization

```python
from mlxtend.plotting import plot_decision_regions

plt.scatter(train_X[0, :], train_X[1, :], c=train_y[0, :], s=40, cmap=plt.cm.Spectral);
plot_decision_regions(train_X.T, train_y[0, :], clf=model_noreg,zoom_factor=1.)

plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.title('w/o Regularization')
plt.xlim(-0.75, 0.5)
plt.ylim(-1, 0.75)
plt.show()
```

# Design a neural network w/ regularization

```python
from keras import models
from keras import layers
from keras.regularizers import l2

model_L2 = models.Sequential()
model_L2.add(layers.Dense(128, input_dim=2, activation='relu', kernel_regularizer=l2(0.01)))
model_L2.add(layers.Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
model_L2.add(layers.Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
model_L2.add(layers.Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
model_L2.add(layers.Dense(1, activation='sigmoid', kernel_regularizer=l2(0.01)))
model_L2.compile(optimizer='rmsprop', 
                      loss='binary_crossentropy', 
                      metrics=['accuracy'])

history_L2 = model_L2.fit(train_X.T, train_y.T, epochs=1000, batch_size=32, validation_data=(test_X.T,test_y.T))
```

```python
model_Dropout = models.Sequential()

model_Dropout.add(layers.Dense(128, input_dim=2, activation='relu'))
model_Dropout.add(layers.Dropout(0.5))
model_Dropout.add(layers.Dense(128, activation='relu'))
model_Dropout.add(layers.Dropout(0.5))
model_Dropout.add(layers.Dense(128, activation='relu'))
model_Dropout.add(layers.Dropout(0.5))
model_Dropout.add(layers.Dense(128, activation='relu'))
model_Dropout.add(layers.Dropout(0.5))
model_Dropout.add(layers.Dense(1, activation='sigmoid'))

model_Dropout.compile(optimizer='rmsprop',
                             loss='binary_crossentropy',
                             metrics=['accuracy'])

history_Dropout = model_Dropout.fit(train_X.T, train_y.T, epochs=1000, batch_size=32, validation_data=(test_X.T,test_y.T))
```

# Draw decision boundaries w/ regularization

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,7))

ax1.scatter(train_X[0, :], train_X[1, :], c=train_y[0, :], s=40, cmap=plt.cm.Spectral);
plot_decision_regions(train_X.T, train_y[0, :], clf=model_L2,zoom_factor=1.,ax=ax1)
ax1.set_xlabel('feature 1')
ax1.set_ylabel('feature 2')
ax1.set_title('L2 Regularization')
ax1.set_xlim(-0.75, 0.5)
ax1.set_ylim(-1, 0.75)

ax2.scatter(train_X[0, :], train_X[1, :], c=train_y[0, :], s=40, cmap=plt.cm.Spectral);
plot_decision_regions(train_X.T, train_y[0, :], clf=model_Dropout,zoom_factor=1.,ax=ax2)
ax2.set_xlabel('feature 1')
ax2.set_ylabel('feature 2')
ax2.set_title('Dropout Regularization')
ax2.set_xlim(-0.75, 0.5)
ax2.set_ylim(-1, 0.75)
plt.show();
```

# Compare the accuracy between the models on the test dataset

```python
# same could be done with ["val_accuracy"], but it's nicer to look at this way in my opinion.
plt.plot(history_Dropout.history["val_loss"])
plt.plot(history_L2.history["val_loss"])
plt.plot(history.history["val_loss"])
plt.legend(["Dropout reg","L2 reg","No regularization"])
plt.title("Loss comparison on test data")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.show();
```
